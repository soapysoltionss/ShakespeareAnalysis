```python
import re
frequency = {}
dt = open('pg100.txt', 'r',encoding='utf-8')
text_string = dt.read().lower()
match_pattern = re.findall(r'\b[a-z]{3,15}\b', text_string)
 
for word in match_pattern:
    count = frequency.get(word,0)
    frequency[word] = count + 1
most_frequent = dict(sorted(frequency.items(), key=lambda elem: elem[1], reverse=True))
most_frequent_count = list(most_frequent.keys())
 
for i in range(0,100):
    print(most_frequent_count[i],most_frequent[most_frequent_count[i]])
```
